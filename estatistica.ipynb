```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
plt.rcParams["figure.figsize"] = (14, 5)
```

```python
data = pd.read_csv("datasets/playlist.csv")
```

```python
fig, ax = plt.subplots()

ordinal_times = np.linspace(data.video_publ.min(), data.video_publ.max(), 13, dtype=np.int)
times = [datetime.fromordinal(i).strftime("%m/%y") for i in ordinal_times]
plt.setp(ax, xticks=ordinal_times, xticklabels=times)
ax.grid(True)

sns.violinplot(data.video_publ, inner=None, ax=ax)
sns.swarmplot(data.video_publ, color="black", ax=ax)
```

```python
fig, ax = plt.subplots()

ordinal_times = np.linspace(data.video_publ.min(), data.video_publ.max(), 13, dtype=np.int)
times = [datetime.fromordinal(i).strftime("%m/%y") for i in ordinal_times]
plt.setp(ax, xticks=ordinal_times, xticklabels=times)
ax.grid(True)

sns.kdeplot(data.video_publ, data.comments, ax=ax)
ax.scatter(data.video_publ, data.comments)
```
